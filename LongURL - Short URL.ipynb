# LONG URL- SHORT URL

## Business Problem:
In this Project, We are considering real world problem of URL. Generally URLs are very long to read and write also. It take lot of time to type a complete URL during heavy workload.

## Solution:
So I have come up with the solution by assigning Short URL to original Long URL. Here we will be using Procedural way of implementation of ShortURL system.
